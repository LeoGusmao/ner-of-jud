```python
def textoJSON(pathJSON):
    textos = []
    with open(pathJson, 'rb') as f:
        dados = f.readlines()
        for databyte in dados:
            datadecode = databyte.decode(encoding='utf-8')
            datajson = json.loads(datadecode)
            texto = datajson['texto']
            textos.append(texto)
    return(textos)
```

# Integrando o Spacy e Scikitlearn

```python
import pandas as pd
pd.set_option('max_colwidth', 250)
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def preprocessamento(texto):
    lista_final = []
    tokens = nlp(texto)
    for token in tokens:
        if(token.is_alpha and not token.is_stop):
            lista_final.append(token.lemma_.lower())
    return ' '.join(lista_final)
```
